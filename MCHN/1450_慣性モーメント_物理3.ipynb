# 回転に関する運動方程式

## 慣性モーメント
moment of inertia


### (慣性)質量
- mass 
- $m \vec{a}= m\ \dfrac{{\rm d^2}\vec{r}}{{\rm d}t^2} = \vec{F} \to \vec{a} = \dfrac{\vec{F}}{m}$
  - $m$が物体の **動かしにくさ** を表す量に見えますか?

## 運動量 $\vec{p}$

- $\vec{p}=m \vec{v}　\to \vec{v} = \dfrac{\vec{p}}{m}$
  - $m$が物体の **動かしにくさ** を表す量に見えますか?
- $\dfrac{{\rm d}}{{\rm d}t}\vec{p}=m \dot{\vec{v}} = m \vec{a}$

回転運動に対して, 動かしにくさ, **回転しにくさ** を表す量を
**慣性モーメント** といいます.

ここから先は, 教科書 p49 図1-34 を利用して考えていきます.

位置ベクトル $\vec{r}$ に質量 $m$ が置かれ, $\vec{r}$ の始点を中心に
$x-y$ 平面を回転している物体を _3次元で_ 考えます.


\begin{align}
\vec{r}
= \left(
  \begin{array}{c}
    x \\
    y \\
    z
  \end{array}
\right)
= \left(
  \begin{array}{c}
    r \cos \theta \\
    r \sin \theta\\
    0
  \end{array}
\right)
= r \left(
  \begin{array}{c}
    \cos \theta \\
    \sin \theta\\
    0
  \end{array}
\right)
\end{align}

$\left| \vec{r} \right| =r =一定$ を考えているので, 速度ベクトル $\vec{v}$  は
\begin{align}
\vec{v} = \dot{\vec{r}} 
= \left(
  \begin{array}{c}
    -r \dot{\theta}\sin \theta \\
    r \dot{\theta}\cos \theta\\
    0
  \end{array}
\right)
=   r \dot{\theta}
 \left(
  \begin{array}{c}
    -\sin \theta \\
    \cos \theta\\
    0
  \end{array}
  \right)
\end{align}

角運動量 $\vec{L}$ は
\begin{align}
  \vec{L}
  &= \vec{r} \times \vec{p} = \vec{r} \times \left( m\vec{v} \right)\\
  &= m r^2 \theta \left(
  \begin{array}{c}
    \cos \theta \\
    \sin \theta\\
    0
  \end{array}
\right)\times  \left(
  \begin{array}{c}
    -\sin \theta \\
    \cos \theta\\
    0
  \end{array}
  \right)
= m r^2 \dot{\theta}  \left(
  \begin{array}{c}
    0\\
    0\\
    1
  \end{array}
  \right)\\[3mm]
  %
\vec{L} &= m r^2 \vec{\omega}, \qquad \vec{\omega} = \left(
  \begin{array}{c}
    0\\
    0\\
    \omega
  \end{array}
  \right), \quad \omega = \dot{\theta}
\end{align}

この式を利用し, 運動量と角運動量を併記すると
\begin{align}
&\begin{cases}
\vec{p} &= m \ \vec{v}\\
\vec{L} &= mr^2 \ \vec{\omega}
\end{cases}\\
&\quad L_z = mr^2 \omega
\end{align}


角運動量 $\vec{L}$ の左辺に現れる $mr^2$ は, 運動量 $\vec{p}$ の質量に類似しています.

そこで
\begin{align}
I = m r^2
\end{align}
とおいて, これを[等速円運動を行う物体の] **慣性モーメント** と呼びます.

運動量 $\vec{p}$ の時間変化を考えることで, 運動方程式 $m \vec{a}= \vec{F}$ が得られました.

同じように, 角運動量 $\vec{L}$ の時間変化を考えると, 回転の運動方程式
\begin{align}
\dfrac{{\rm d}\vec{L}}{{\rm d}t} = \vec{r} \times \vec{F} = \vec{M}
\end{align}
が得られました.

$\require{color}$
[等速円運動を行う物体の] 角運動量 $\vec{L}$ をこの回転の運動方程式に形式に当てはめ, $z$ 成分のみに注目すると
\begin{align}
mr^2 \ddot{\theta} &= M_z\\
mr^2 \ddot{\theta} &= M_z\\[3mm]
\textcolor{green}{mr^2} \dfrac{{\rm d} \omega}{{\rm d}t} &= M_z
\end{align}

この形を運動方程式と比較すると
\begin{align}
\textcolor{green}{m} \dfrac{{\rm d} {v}}{{\rm d}t} &= F
\end{align}

この比較を通じても, 回転のしにくさを表す **慣性モーメント** $I$ を
\begin{align}
I = m r^2
\end{align}
とおくと, 回転の運動方程式は形式的に
\begin{align}
\textcolor{green}{I} \dfrac{{\rm d} \omega}{{\rm d}t} = M_z
\end{align}
と記述することができます.

これは運動方程式
\begin{align}
\textcolor{green}{m} \dfrac{{\rm d} {v}}{{\rm d}t} &= F
\end{align}
と類似性(アナロジー, analogy) を持っています．

### 慣性モーメントによる角運動量

$\require{color}$
\begin{align}
\vec{L}
&= \textcolor{green}{m r^2} \vec{\omega}\\
&= \textcolor{green}{I} \vec{\omega}
\end{align}
特に, 教科書 p50 図1-34 の様な固定した $z$軸 での角運動量は
\begin{align}
{L}_z
&= \textcolor{green}{I} \omega
\end{align}
となります.
