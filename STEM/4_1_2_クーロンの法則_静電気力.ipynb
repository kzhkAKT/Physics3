### $Q$ と $q$ を結ぶ直線上の引力と斥力

\begin{align}
F &= k \dfrac{Q \ q }{r^2}=
\begin{cases}
Qq >0 &  \text{電荷同符号は斥力}
\begin{cases}
Q>0 \ \text{and}\  q>0\\
Q<0 \ \text{and}\  q<0
\end{cases}\\[2mm]
%
Qq < 0 &  \text{電荷異符号は斥力}
\begin{cases}
Q>0 \ \text{and}\  q<0\\
Q<0 \ \text{and}\  q>0
\end{cases}
\end{cases}
\end{align}

\begin{align}
k = k_e = \dfrac{1}{4\pi \epsilon_0} \approx {\rm 8.988\times 10^9\  N⋅m^2⋅C^{-2}} \sim {\rm 9.0\times 10^9\  N⋅m^2⋅C^{-2}}
\end{align}

$\epsilon_0$ は真空の誘電率  
$(\epsilon_0{\rm = 8.8541878128(13)\times 10^{-12}\  F⋅m^{-1}})$

#### クーロン力

距離 $r_i$ 離れた 外部点電荷 $q_i$ から,  
観察している点電荷 $Q$ に働く力
\begin{align}
\vec{F}_{i} = \left| \vec{F}_{i}  \right| \vec{e}_{i}
\ & \  
\begin{cases}
\left| \vec{F}_{i}  \right| = k \dfrac{|Q||q_i|}{r_i^2}\\[2mm]
 \vec{e}_{i}: \ \text{$Q$ と $q_i$ を結ぶ直線上での引力($Qq_i<0$)/斥力($Qq_i>0$)}
\end{cases}
\end{align}

### 力の合成

\begin{align}
\vec{F} &= \sum_{i=1}^{N} \vec{F}_{i}\\
&= \vec{F}_{1} + \vec{F}_{2} + \cdots + \vec{F}_{i} + \cdots + \vec{F}_{N} \\[3mm]
\end{align}

#### クーロン力の重ねあわせ

- 注目する電荷と外部電荷との積で引力/斥力の判断
- 一つ一つのクーロン力をベクトル$(=\left| \vec{F}_{i}  \right| \vec{e}_{i})$ で表示
- クーロン力の合成を行い, 合力 $\vec{F}=\sum \vec{F}_i$ を求める
  - ベクトルは座標に持ち込み成分で調べる
  - 幾何学的/図形的な関係を用いて調べる

- 力で表現されるので, 重力や万有引力とも合力を調べることができる.
